```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(input_dim=58,init="random_uniform",output_dim=18,activation="relu"))
```

```python
model.add(Dense(init="random_uniform",output_dim=13))
```

```python
model.add(Dense(init="random_uniform",output_dim=3))
```

```python
model.add(Dense(init="random_uniform",output_dim=1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mse'])
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=8)
```

```python
model.save("project_class.h5")
```
